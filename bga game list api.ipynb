```python
import requests
from bs4 import BeautifulSoup
import json
import time
import random
```

```python
total_results = []

# Loop through from pages 1 to 6
for page_num in range(0,2050,25):
    # Build the URL and download the results
    url = "https://api.boardgameatlas.com/api/search?&skip=" + str(page_num) + "&client_id=W0AQGbjlZE"
    print("Downloading", url)
    response = requests.get(url)
    time.sleep(random.random()*5)
    data = response.json()
    total_results = total_results + data['games']

print("We have", len(total_results), "total games")
```

```python
#for i in total_results:
    #print(i['name'],i['price'])
    
import pandas as pd
#import pandas as pd
df = pd.DataFrame()  # initialize dataframe

# loop through each post retrieved from GET request
for items in total_results:
    # append relevant data to dataframe
    df = df.append({
       'name': items['name'],
       'price':items['price'],
       'price_ca':items['price_ca'] ,
       'price_uk':items['price_uk'],
       'price_au':items['price_au'],
       'min_players':items['min_players'] ,
       'max_players':items['max_players'],
       'min_playtime':items['min_playtime'],
       'max_playtime':items[ 'max_playtime'],
       'min_age':items['min_age'],
       'description':items['description'],
       'num_user_ratings':items['num_user_ratings'],
       'average_user_rating':items['average_user_rating']
    }, ignore_index=True)

    
```

```python
order = ['name', 'average_user_rating','num_user_ratings','min_players', 'max_players', 'min_playtime', 'max_playtime', 'price_au', 'price_ca','price_uk']
df = df[order]
df.to_csv("bga_game_items.csv",index=False)
```
