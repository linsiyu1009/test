```python
# # load libraries

# To support both python 2 and python 3
from __future__ import division, print_function, unicode_literals

# Common imports
import numpy as np
import os
import pandas as pd

#show all rows and columns
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# load data
luding_data = pd.read_csv('luding_data.csv', header = 0) #change csv name
```

```python
# Keep a copy of original data in case of changes made to dataframe
data = luding_data.copy() #make a copy and do not need to change anything from here on 
```

```python
data.head()
```

```python
#find out the proportion of missing values in each variable
data.isnull().sum()/len(data)
```

```python
# drop all columns with 100% missing values
# data = data.dropna(axis='columns', how ='all')

#drop columns having more than 70% null values
data = data.drop(data.loc[:,list((100*(data.isnull().sum()/len(data.index))>70))].columns, 1)
```

```python
#find out the proportion of missing values in each variable
data.isnull().sum()/len(data)
```
