```python
# 使用API对象获取你的时间轴上的微博，并把结果存在一个叫做public_tweets的变量中 
public_tweets = api.home_timeline()   
# 遍历所拉取的全部微博 
for tweet in public_tweets:     
# 打印存在微博对象中的text字段    
    print(tweet.text)
```
